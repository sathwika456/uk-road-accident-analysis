# UK Road Accident Data Analysis
This notebook demonstrates a sample workflow of loading and exploring the UK road accident datasets using pandas and seaborn.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the accident dataset
accident_df = pd.read_csv('datasets/Road Accident Data.csv')
accident_df.head()
```

```python
# Summary statistics
accident_df.describe(include='all')
```

```python
# Plot number of casualties by weather conditions
plt.figure(figsize=(10,6))
sns.boxplot(data=accident_df, x='Weather_Conditions', y='Number_of_Casualties')
plt.xticks(rotation=45)
plt.title('Number of Casualties by Weather Conditions')
plt.tight_layout()
plt.show()
```

```python
# Check for missing values
accident_df.isnull().sum().sort_values(ascending=False)
```
